```python
# Change the types of dataset
data['User ID']=data['User ID'].astype(int)
data['Age']=data['Age'].astype(int)
data['EstimatedSalary']=data['EstimatedSalary'].astype(int)
data['Purchased']=data['Purchased'].astype(int)
```
